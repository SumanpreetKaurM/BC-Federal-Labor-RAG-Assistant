# Week 1 – Baseline Keyword Search

## Objective

The purpose of this notebook is to implement a simple keyword-based search system that simulates how traditional text searching (similar to Ctrl + F) works.

This baseline implementation will be used as a comparison against the AI-powered Retrieval-Augmented Generation (RAG) assistant that will be developed in later weeks.

The notebook demonstrates the limitations of exact keyword matching when searching labour law documents.

```python
# week1_baseline_search.ipynb
# Purpose: Create an un-optimized word match baseline to compare against our future AI system.

def local_keyword_matcher(keyword, document_text):
    """Scans clean text blocks line-by-line for exact word string matches."""
    lines = document_text.split('\n')
    results = [line.strip() for line in lines if keyword.lower() in line.lower()]
    return results[:3]  # Return the top 3 matching sentences found


# Simulated cross-jurisdiction legal text snippets
sample_jurisdiction_data = """
[BC Provincial Law] ESA Section 49.1: Eligible employees get 5 permanent paid sick leaves per calendar year.
[Federal Canada Law] CLC Section 239: Federally regulated employees can accumulate up to 10 paid sick days.
[BC Provincial Law] ESA Part 4 Section 40: Overtime is calculated at 1.5x regular pay after 8 hours in a workday.
[Federal Canada Law] CLC Part III Section 174: Overtime is paid at 1.5x the regular rate for hours worked beyond standard limits.
"""


# Execute a keyword extraction test
test_keyword = "sick leaves"
baseline_output = local_keyword_matcher(test_keyword, sample_jurisdiction_data)

print(f"--- Raw Keyword Match Results for: '{test_keyword}' ---")

for index, line_match in enumerate(baseline_output, 1):
    print(f"{index}. {line_match}")
```

## Results

The keyword **"sick leaves"** successfully matched the British Columbia employment law sentence.

The federal sentence was not returned because it contains the words **"paid sick days"** instead of **"paid sick leaves."**

This demonstrates one limitation of traditional keyword search: it relies on exact word matching and cannot understand that different words may have the same meaning.

This limitation motivates the use of an AI-powered RAG system, which will be implemented in later stages of the project.
